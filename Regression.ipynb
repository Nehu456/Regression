 What is Heteroscedasticity?
In Multiple Linear Regression, heteroscedasticity occurs when the variance of the residuals (errors) is not constant across all levels of the independent variables.

In other words:

The spread of the errors changes as the predicted values (or X values) increase.

Think of a funnel shape when plotting residuals ‚Äî either narrowing or widening.

This pattern indicates heteroscedasticity.

 How It Affects the Model:
Inefficient Estimates

Coefficients (slopes) may still be unbiased, but they are not the most efficient.

Standard errors are incorrect ‚Üí affects confidence intervals and p-values.

Hypothesis Testing Becomes Unreliable

T-tests and F-tests may give misleading results.

You might think a variable is significant when it‚Äôs not (or vice versa).

Model Interpretability Reduces

Makes predictions less trustworthy, especially at extreme values of X.



Answer 12

1. Check VIF (Variance Inflation Factor)
VIF > 5 (or 10 in relaxed cases) indicates potential multicollinearity.

Drop variables with high VIF values.

2. Remove or Combine Correlated Variables
Drop one of the correlated variables.

Or create a composite variable (e.g., average or principal component).

3. Use Principal Component Analysis (PCA)
PCA converts correlated variables into a smaller set of uncorrelated components.

Especially useful when you have many predictors.

4. Regularization Techniques (Ridge or Lasso Regression)
These techniques penalize large coefficients, helping manage multicollinearity:

Ridge Regression: Shrinks coefficients but keeps all variables.

Lasso Regression: Can shrink some coefficients to zero (acts like variable selection too).

5. Domain Knowledge
Sometimes, the best fix is to use your understanding of the data to decide which variables are more meaningful and should stay.

Answer 13

1. One-Hot Encoding (Dummy Variables)
Creates a new binary column for each category (1 = yes, 0 = no).

Use when categories are nominal (no natural order), like:
Color: Red, Blue, Green ‚Üí [Color_Red, Color_Blue, Color_Green]

2. Label Encoding
 Assigns a unique number to each category.

Best for ordinal data (where order matters), like:
Size: Small = 0, Medium = 1, Large = 2

3. Ordinal Encoding
 Similar to label encoding, but you manually define the order.

 More reliable when you know the ranking of categories.

 4. Binary Encoding / Hashing (Advanced)
 Good for high-cardinality features (like 100s of unique categories).

 Uses fewer columns than one-hot but keeps meaningful info.

Use category_encoders library in Python.

5. Frequency or Count Encoding
Replaces each category with how often it appears.

 Good for large datasets, but can leak target info if not careful.

Answer 17


 How the Intercept Provides Context:
1. It Anchors the Line (or Plane)
The intercept sets the starting point of the prediction line.

Without it, the model assumes the line goes through the origin (0, 0), which is rarely accurate.

2. It Reflects the Baseline Value of Y
It gives the expected value of the dependent variable when all predictors are at their minimum (or 0).

3. Helps Understand Effects Relative to a Reference Group (in Categorical Variables)
In multiple regression with categorical variables:

The intercept often represents the predicted Y for the baseline (reference) category when all numerical variables are 0.

What is R¬≤?
R¬≤ tells us the proportion of variance in the dependent variable that is explained by the independent variables.

Ranges from 0 to 1:

0 = model explains nothing

1 = model explains everything

 Limitations of Using R¬≤ Alone:
1.  R¬≤ Always Increases When You Add More Predictors
Even if the new variables are not meaningful, R¬≤ will still go up.

That‚Äôs why we often use Adjusted R¬≤ in Multiple Linear Regression to penalize irrelevant predictors.

2.  Doesn‚Äôt Indicate Causation
A high R¬≤ doesn‚Äôt mean X causes Y ‚Äî it only shows a relationship.

Example: Ice cream sales and shark attacks may both go up in summer ‚Üí high R¬≤ but no causal link!

3.  Doesn‚Äôt Tell You if the Model is Biased
R¬≤ doesn't reveal whether:

Residuals are biased

Assumptions like linearity or homoscedasticity are violated

So, a model can have high R¬≤ but poor predictive power on new data.

Answer 22

 Why Is Scaling Important in Multiple Linear Regression?
1. Makes Coefficients Comparable
If your predictors are on different scales (e.g., age in years vs. income in lakhs), their coefficients can't be compared directly.

Scaling allows you to interpret coefficients in standard units ‚Äî useful for determining which variable has more influence.



2. Improves Numerical Stability
Large differences in magnitude between variables can cause computational issues.

Scaling reduces the risk of rounding errors and improves the accuracy of coefficient estimates.

3. Essential for Regularization Techniques
If you're using:

Ridge Regression

Lasso Regression

ElasticNet

‚Ä¶then scaling is mandatory. These methods penalize coefficients based on their magnitude, so inconsistent scales would unfairly penalize some features more than others.

Answer 29

1. Visual Inspection (Plotting the Fit)
Plot the actual data vs. the polynomial regression curve.

Helps you visually assess if the model is underfitting or overfitting.

Use this alongside other metrics ‚Äî visuals alone can be misleading.

2. Residual Plots
Plot residuals (errors) vs. predicted values.

A good model should have random scatter ‚Äî no patterns.

If residuals show curves ‚Üí model is underfitting.

If residuals are all over the place ‚Üí possible overfitting.


3.  R¬≤ and Adjusted R¬≤
R¬≤ shows how much variance is explained by the model.

Adjusted R¬≤ is better for comparing models with different numbers of terms ‚Äî it penalizes adding unnecessary degrees.

Choose the degree where Adjusted R¬≤ stops increasing significantly.

4. Cross-Validation (like k-Fold CV)
Split data into training and validation sets multiple times.

Calculate the average prediction error (e.g., Mean Squared Error) for each degree.

Choose the degree with the lowest average validation error.

This is one of the most robust methods to avoid overfitting!



Answer 30

1. Understand the Shape of the Relationship
Polynomial regression can model curved relationships (not just straight lines).

Visualizing helps you see the actual curve and how well it captures your data trend.

A degree 2 (parabola) or degree 3 (S-curve) is hard to grasp without plotting it.

2. Detect Underfitting or Overfitting
A low-degree polynomial may not follow the curve of your data ‚Üí underfitting.

A high-degree polynomial may twist and turn too much ‚Üí overfitting.

A plot clearly shows whether your model is too simple or too complex.

Example:

If the model curve swings wildly between points:  likely overfitting.

If the model misses the shape of the data: likely underfitting.

3.  Check Residual Patterns
Plotting residuals vs. predicted values shows if there‚Äôs a pattern.

If residuals curve ‚Üí your current degree might not be enough.

If residuals are random ‚Üí model is likely well-fitted.

4. Compare Different Degrees Easily
Visual plots help you compare how degree 2 vs. degree 3 vs. degree 5 fits the data.

You can visually pick the simplest degree that captures the pattern well.
